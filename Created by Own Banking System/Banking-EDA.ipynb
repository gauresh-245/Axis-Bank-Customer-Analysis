```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('Banking.csv')
df.head()
```

```python
# ( function that gives you a quick statistical summary )
# It shows you mean,std,max ,count 

df.describe()
```

```python
#Statment:  Knowing which income group your largest customer base belongs to helps:


bins = [0,100000,300000, float('inf')]
labels = ['Low','Med','High']

df['Income Band'] = pd.cut(df['Estimated Income'], bins=bins,labels=labels,right=False)

df['Income Band'].value_counts().plot(kind='bar')
```

```python
# Analyze the frequency of values in each categorical column.

# Compare that distribution by gender.

categorial_cols = [
    "BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality",
    "Occupation", "Fee Structure", "Loyalty Classification", 
    "Properties Owned", "Risk Weighting", "Income Band"
]

for i, predictor in enumerate(categorial_cols):
    plt.figure(i)
    sns.countplot(data=df, x=predictor, hue="GenderId")
    plt.tight_layout()

plt.show()
```

```python
# Histplot of value counts for different Occupations

# You want to generate histograms of value counts for all categorical columns except "Occupation"


for col in categorial_cols:
    if col == "Occupation":
        continue
    plt.figure(figsize=(8,4))
    sns.histplot(df[col])
    plt.title(f"Histogram of {col} count")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

Numerical Analysis

```python
numerical_cols = ['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']

plt.figure(figsize=(15,10))

for i,col in enumerate(numerical_cols):
    plt.subplot(4,3,i+1)
    sns.histplot(df[col],kde = True)
    plt.title(col)

plt.show()
```

```python
numerical_cols = ['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']


correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(12,12))

sns.heatmap(correlation_matrix,annot=True,cmap="crest",fmt=".2f")
plt.title("correlation Matrix")
plt.show()
```
